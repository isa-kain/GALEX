```python
import numpy as np
import pandas as pd
import re
import warnings
from tqdm import tqdm

import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import ListedColormap, LinearSegmentedColormap

from astropy.io import fits
import astropy.units as u
from astropy.table import Table, vstack, hstack, join
```

```python
def absolutemag(m, d):
    '''
    Inputs:
    m: AB apparent magnitude
    d: distance (pc)
    
    Returns:
    M: absolute magnitude
    '''
    
    return m - 5 * ( np.log10(d) - 1)
```

## Read in data

```python
## Dataset xmatched between IUE/GALEX, cuts on spectral type (FGK) and object type ("normal" stars)
data = Table.read('/Users/isabelkain/Desktop/GALEX/data/dataset.ecsv')

## Dataset xmatched between IUE/GALEX, cuts on spectral type (FGK)
dataFGK = Table.read('/Users/isabelkain/Desktop/GALEX/data/dataset_FGK_alltypes.ecsv')

## Dataset xmatched between IUE/GALEX, cuts on object type (stars)
datastars = Table.read('/Users/isabelkain/Desktop/GALEX/data/dataset_stars.ecsv')
```

```python
## Convert FUV, NUV, V, and B apparent mags to absolute mags

data['FUVabsmag'] = absolutemag(data['FUVmag'], data['Distance_distance'])
data['NUVabsmag'] = absolutemag(data['NUVmag'], data['Distance_distance'])
data['Vabsmag'] = absolutemag(data['V'], data['Distance_distance'])

dataFGK['FUVabsmag'] = absolutemag(dataFGK['FUVmag'], dataFGK['Distance_distance'])
dataFGK['NUVabsmag'] = absolutemag(dataFGK['NUVmag'], dataFGK['Distance_distance'])
dataFGK['Vabsmag'] = absolutemag(dataFGK['V'], dataFGK['Distance_distance'])

# datastars['FUVabsmag'] = absolutemag(datastars['FUVmag'], datastars['Distance_distance'])
# datastars['NUVabsmag'] = absolutemag(datastars['NUVmag'], datastars['Distance_distance'])
# datastars['Vabsmag'] = absolutemag(datastars['V'], datastars['Distance_distance'])
```

### Object distributions

```python
## Count occurence rates of object types in sample

cats = np.unique(data['main_type'])
counts = np.zeros(len(cats))

for i in range(len(cats)):
    counts[i] = np.sum( data['main_type'] == cats[i] )


## Ignore categories with <2 members, sort descending

# cnt = counts >= 10
# cats = cats[cnt]
# counts = counts[cnt]

order = np.argsort(counts)
cats = cats[order]
counts = counts[order]

## Plot hbar

plt.tight_layout()

plt.figure(figsize=(4,6))
plt.barh(cats, counts, edgecolor='k')
plt.xlabel('Counts', fontsize=12, c='k');
plt.title(f'SIMBAD object types in sample\nCuts on spectral type (FGK), object type ("normal" stars)\nSample size: {len(data)}', fontsize=14, c='k')

plt.xscale('log')
plt.yticks(fontsize=12, c='k');
plt.xticks(fontsize=12, c='k');
plt.ylim([-1, len(cats)])
```

## Normal vs. abnormal comparison plots (Graeme)

```python
ot = dataFGK['main_type']
otypes = np.unique( ot[(ot!='Star')&(ot!='PM*')] )

## Drop object types with <5 members
trim = [np.sum(ot==otypes[i])>=5 for i in range(len(otypes))]
otypes = otypes[trim]
print(otypes)

normal = dataFGK[dataFGK['main_type']=='Star']
propm = dataFGK[dataFGK['main_type']=='PM*']
```

### C-M , FUV vs FUV-B

```python
for obj in otypes:
    
    print(f'{obj}')
    objgroup = dataFGK[dataFGK['main_type']==obj]

    plt.figure()

    plt.scatter(normal['FUVmag'] - normal['B'],
                normal['FUVabsmag'],
                ec='k', c='None', marker='o',
                label='Stars')

    plt.scatter(propm['FUVmag'] - propm['B'],
                propm['FUVabsmag'],
                ec='k', c='None', marker='^',
                label='High PM')
    
    plt.scatter(objgroup['FUVmag'] - objgroup['B'],
                objgroup['FUVabsmag'],
                ec='k', c='k', marker='o',
                label=f'{obj} ({np.sum(ot==obj)})')
    
    plt.xlabel('FUV - V')
    plt.ylabel('FUV abs mag')
    plt.legend(loc='upper left')
    
    plt.savefig(f'/Users/isabelkain/Desktop/GALEX/ColorMagnitudePlots/FUV_FUV-B/FUV_FUV-B_{obj}.png')
```
