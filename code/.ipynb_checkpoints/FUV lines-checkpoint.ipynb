```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from astropy.io import fits
import astropy.units as u
from astropy.table import Table, vstack, hstack
import re

from astroquery.simbad import Simbad
from astroquery.xmatch import XMatch
from astroquery.vizier import Vizier
from astroquery.atomic import AtomicLineList
from astroquery.nist import Nist
from astroquery.hitran import Hitran
```

Figuring out which lines 1) might be present in spectra, based on  the composition of chromospheres, common transitions, and abundances; and 2) will be bright enough to identify in spectra. Relative intensities of each line are given by NIST, which places big caveats on their reliability. https://physics.nist.gov/PhysRefData/ASD/Html/lineshelp.html#OUTRELINT

Chromospheres are composed of hydrogen, helium, sodium, magnesium, calcium, and iron.

GALEX operates in two bands: the FUV channel (1350 - 1750 A) and NUV channel (1750 - 2800 A). https://asd.gsfc.nasa.gov/archive/galex/Documents/instrument_summary.html


```python
## Querying NIST

tableCaII = Nist.query(135*u.nm, 280*u.nm, output_order='wavelength', linename="Ca II")

rel = np.array(tableCaII['Rel.'])
# relCaII = np.array([float(re.split('[a-zA-Z:*?]', el)[0]) for el in rel])

## Strip NIST identifiers on relative intensities, and throw out lines marked as sketch

relCaII = []
dropix = []
for i, el in enumerate(rel):
    
    val = re.split('[a-zA-Z:*?-]', el)

    try:
        v = float( val[0] )
        relCaII.append( v )
    except:
#         print(i, val)
        dropix.append(i)
        relCaII.append(0)
        continue

    
relCaII = np.array(relCaII)

FUVloc = tableCaII['Observed']<175
NUVloc = tableCaII['Observed']>=175

FUVloc[dropix] = False
NUVloc[dropix] = False

print('Table entries:', len(tableCaII))
print('Viable lines:', len(relCaII))

## Plot lines falling into GALEX's FUV and NUV bands

plt.scatter( tableCaII['Observed'][FUVloc],  relCaII[FUVloc], color='goldenrod', label='GALEX FUV channel')
plt.scatter( tableCaII['Observed'][NUVloc],  relCaII[NUVloc], color='darkorchid', label='GALEX NUV channel')
plt.legend()
# plt.yscale('log')

plt.xlabel('Wavelength (nm)')
plt.ylabel('NIST relative intensity')
plt.title('Ca II lines in GALEX filter bands')
```
