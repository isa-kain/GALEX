## Load FITS data

Each spectrograph has a prime and a redundant camera.
    The Long-Wavelength Prime (LWP) and Short-Wavelength Prime (SWP)
    cameras are the standard detectors (the LWR (Long-Wavelength Redundant)
    was used before Oct, 1983).
    
    https://archive.stsci.edu/iue/file_formats.html
    
   Camera     (FULL) High Dispersion (PARTIAL)          Low Dispersion

  LWP       1845-2980 A      2980-3230 A               1910-3300 A
  LWR       1845-2980 A      2980-3230 A               1910-3300 A
  SWP       1145-1930 A      1930-2198 A               1150-1975 A

Order of HDUL[1] bintable fields for MX files:
    
[1] Order number, one 8-bit byte.

[2] Number of extracted points n, one 16-bit integer.

[3] Starting wavelength, one double-precision floating point value.

[4] Starting pixel at starting wavelength, one 16-bit integer.

[5] Wavelength increment, one double-precision floating point value.

[6] Slit height in pixels, one single-precision floating point number.

[7] Line number for found centroid of spectrum, one single-precision floating point number.

[8] Net flux spectrum, 768 single-precision floating point numbers with n extracted data points.

[9] Background flux spectrum, 768 single-precision floating point numbers with n extracted data points.

[10] Noise vector, 768 single-precision floating point numbers with n extracted data points.

[11] $\nu$ flags as n 16-bit integers stored in two's complement form.

[12] Ripple-corrected net flux spectrum, 768 single-precision floating with n extracted data points.

[13] Absolutely-calibrated, ripple-corrected net flux spectrum, 768 single-precision floating point numbers. with n 
extracted data points.

[14] Start pixel for background fit, one 16-bit integer number. *

[15] End pixel for background fit, one 16-bit integer number. *

[16] Chebyshev scale factor, one single-precision floating point number. *

[17] Chebyshev polynomial coefficients for global background correction, 7 single-precision floating point numbers. *

```python
# from: https://github.com/spacetelescope/MASTDataDelivery/blob/60aa8468e55d0d76e1245c3a956a8429e42f4107/get_data_iue.py

def calculate_cut_wl(camera, order, aperture):
    """
    Calculate the cut wavelength for an order following Solano's paper.
    :param camera: The camera used.
    :type camera: str
    :param order: The order number.
    :type order: float
    :param aperture: The aperture used.
    :type aperture: str
    :returns: float -- The cut wavelength for this order.
    """
    # NOTE: When I tried to use it I did not get cut wavelengths that fell
    # within the order's range, either the equation is wrong or I am applying it
    # wrong.
    if camera == "LWP":
        if order >= 77 and order <= 124:
            if aperture == "LARGE":
                cut_wl = -7.9697 + 233257.6280/order
            else:
                cut_wl = -7.7959 + 233382.6450/order
        else:
            # Non-overlap orders, just return a very large wavelength.
            cut_wl = 9E9
    elif camera == "LWR":
        if order >= 76 and order <= 119:
            if aperture == "LARGE":
                cut_wl = -11.3459 + 233737.5903/order
            else:
                cut_wl = -11.2214 + 233876.9950/order
        else:
            cut_wl = 9E9
    else:
        # The only other camera option is SWP.
        if order >= 73 and order <= 120:
            if aperture == "LARGE":
                cut_wl = 24.3952 + 132875.4838/order + 325840.9715/(order*order)
            else:
                cut_wl = 22.2095 + 133293.4862/order + 300351.2209/(order*order)
        else:
            cut_wl = 9E9
            
            
    return cut_wl
```

```python
data = hdul[1].data
camera = hdul[0].header['CAMERA']
aperture   = hdul[0].header['APERTURE']
prev_cut_wl = 9E9

%matplotlib tk

# fig, ax = plt.subplots(15, 1, figsize=(8,25))
# plt.tight_layout()

for i in range(data['ORDER'].size): # data['ORDER'].size

    ## Pull information from FITS headers
    
    order      = data['ORDER'][i] 
    npoints    = data['NPOINTS'][i]
    startwv    = data['WAVELENGTH'][i]
    startpix   = data['STARTPIX'][i]
    deltaw     = data['DELTAW'][i]
    abs_cal    = data['ABS_CAL'][i]
    noise      = data['NOISE'][i]
    netflx     = data['NET'][i]
    
    
    ## Calculate wavelength range and flux

    WVL        = startwv + np.arange(npoints) * deltaw
    FLX        = abs_cal[startpix-1:startpix-1+npoints]
    

    ## Apply wavelength cutoff from m-1 order
    
    if prev_cut_wl < 9E9:
                
        FLX = FLX[ WVL >= prev_cut_wl ]
        WVL = WVL[ WVL >= prev_cut_wl ]
        
        
    ## Calculate wavelength cutoff for current order (m)
    
    cut_wl = calculate_cut_wl(camera, order, aperture)
    
    
    ## Apply wavelength cutoff for current order (m)
    
    if cut_wl < 9E9:
                
        FLX = FLX[ WVL < cut_wl ]
        WVL = WVL[ WVL < cut_wl ]
        
    ## Set current cut wavelength to previous cut wavelength
    print(i, order, 'm cut:', cut_wl, 'WVL range:', WVL[0], WVL[-1])
    prev_cut_wl = np.copy(cut_wl)
    
    ## Report results
        
    plt.plot(WVL, FLX, c=colors[i%10], label=order) ## this actually looks mostly fine. big gaps between orders at longer wavelengths, and the first ~5 orders look wonky, but otherwise quite good
#     ax[i].set_ylabel(f'Order {order}')

# plt.yscale('log')
# plt.xlim(1850,2100)
plt.xlabel(r'Wavelength ($\AA$)', fontsize=14)
plt.ylabel(r'Flux (erg/cm$^2$/s/$\AA$', fontsize=14)
plt.legend(title='Spectrograph order:', markerscale=5, bbox_to_anchor=(1.2, 1.0))
```
